```python
#Assignment 3 - Data Visualization Exercise

#Activity 1 - Line Graph

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([1,2,3,4,5],dtype = 'int')
y = x**2
```

```python
plt.plot(x,y,color='r',linewidth=2)
plt.show()
```

```python
#Activity 2 - scatter plot

x_1 = np.array([1,4,7,9,3,6,8])
x_2 = np.array([12,17,15,19,14,18,13])

plt.scatter(x_1,x_2,marker = '^',color = 'g')
plt.show()
```

```python
#Activity 3 - Bar Graph

x_2 = np.array(['USA','Canada','Germany','France','Russia'])
y_2 = np.array([5200,2200,2400,1600,6100])
colors = ['red','blue','green','yellow','pink']
plt.xlabel("Countries")
plt.ylabel("Missiles")
plt.title("Military Strength")
plt.bar(x_2,y_2,color = colors)
plt.show()
```

```python
#Activity 4 - histogram

x_3 = np.random.normal(180,5,100)

plt.hist(x_3,color = 'teal')
plt.show()
```

```python
#Activity 5 - Count plot
import seaborn as sns

titanic = sns.load_dataset("titanic")
ax = sns.countplot(x="class",data = titanic)
print(ax)
```

```python
#Activity 6 - Heatmap

uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data)
print(ax)
```

```python
#Activity 7 - Box plot

tips = sns.load_dataset("tips")
ax = sns.boxplot(x=tips["total_bill"])
print(ax)
```

```python
#Activity 8 - pairplot

penguins = sns.load_dataset("penguins")
sns.pairplot(penguins)
```

```python
#Activity 9 - joint plot

penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
#Activity 10 - kernal density plot

tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill")
```
